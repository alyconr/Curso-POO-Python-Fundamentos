# üß™ Ejercicio 1: Registro de Empleados
## Enunciado:

* Crea una clase Empleado que contenga:

* Atributo p√∫blico: nombre

* Atributo protegido: _salario

* Atributo privado: __id

* Un contador de empleados como atributo de clase

* M√©todo m√°gico __str__ para imprimir la informaci√≥n del empleado

* M√©todo est√°tico es_valido_id que retorne True si un ID comienza con "EMP"

* Un m√©todo de clase que devuelva cu√°ntos empleados se han creado
* Tambi√©n usa @property y @setter para validar que el salario no sea menor de 1,000,000.

```python
class Empleado: 
    
    contadorEmpleados = 0
    
    def __init__(self, nombre, salario, id):
        if not Empleado.esValidoId(id):
            raise ValueError("El id debe empezar por 'EMP'.")
        
        self.nombre = nombre
        self._salario = salario
        self.__id = id
        Empleado.contadorEmpleados += 1
        
    def  __str__(self):
        return f"Nombre: {self.nombre}, Salario: {self._salario}, Id: {self.__id}"
    
    @staticmethod
    def esValidoId(id):
        return id.startswith("EMP")
    
    @classmethod
    def totalEmpleados(cls):
        return cls.contadorEmpleados
    
    @property 
    def salario(self):
        return self._salario
    
    @salario.setter
    def salario(self, nuevoSalario):
        if nuevoSalario < 1000000:
            raise ValueError("El salario no debe ser menor a 1000000")
        self._salario = nuevoSalario
        
e1 = Empleado("juan", 0, "EMP1")
e1.salario = 1000000

print(e1)
print(f"El total de empleados es: {Empleado.totalEmpleados()}")    
```

# üß™ Ejercicio 2: Cuenta Bancaria
## Enunciado:

* Crea una clase CuentaBancaria que incluya:

* Atributo p√∫blico: titular

* Atributo protegido: _saldo

* Atributo privado: __numero_cuenta

* M√©todo m√°gico __repr__ para representar la cuenta

* M√©todo depositar y retirar

* @property y @setter para acceder/modificar el saldo (no permitir saldo negativo)

* M√©todo de clase desde_string que cree una cuenta desde una cadena con formato: "Luis,1234,500000"

* M√©todo est√°tico es_cuenta_valida que verifique que el n√∫mero de cuenta tenga 4 d√≠gitos

```python
class cuentaBancaria:
    def __init__(self, titular, numeroCuenta, saldo):
        self.titular = titular
        self._saldo = saldo
        self.__numeroCuenta = numeroCuenta
        
    def __repr__(self):
        return f"Titular: {self.titular}, Saldo: {self._saldo}, N√∫mero de cuenta: ****"
    
    def depositar(self, monto):
        if monto > 0:
            self.acceder += monto
            print(f"Dep√≥sito exitoso: {monto}")
        else:
            print("El monto debe ser positivo")

    def retirar(self, monto):
        if monto > 0 and monto <= self.acceder:
            self.acceder -= monto
            print(f"Retiro exitoso: {monto}")
        else:
            print("Fondos insuficientes o monto inv√°lido")
            
    @property 
    def acceder(self):
        return self._saldo
    
    @acceder.setter
    def acceder(self, modificar):
        if modificar < 0:
            raise ValueError("El saldo no debe ser negativo")
        self._saldo = modificar
        
    @classmethod
    def desde_string(cls, cadena):
        partes = cadena.split(',')
        if len(partes) != 3:
            raise ValueError("Formato incorrecto, se esperaba: 'titular,numero_cuenta,saldo'")
        titular, numero_cuenta, saldo = partes
        if not cls.es_cuenta_valida(numero_cuenta):
            raise ValueError("N√∫mero de cuenta inv√°lido")
        return cls(titular, numero_cuenta, float(saldo))
    
    @staticmethod
    def es_cuenta_valida(numeroCuenta):
        return numeroCuenta.isdigit() and len(numeroCuenta) == 4

cuenta = cuentaBancaria.desde_string("Luis,1234,50000")
print(cuenta)

cuenta.depositar(10000)
cuenta.retirar(5000)

print(cuenta.acceder)

cuenta.acceder = 600000
print(cuenta)
```

# üß™ Ejercicio 3: Producto con Descuento
## Enunciado:

* Crea una clase Producto que tenga:

* Atributo p√∫blico: nombre

* Atributo protegido: _precio

* Atributo privado: __codigo

* M√©todo m√°gico __eq__ para comparar productos por su c√≥digo

* M√©todo est√°tico aplicar_descuento(precio, porcentaje) que calcule el nuevo precio

* Atributo de clase impuesto que todos los productos comparten

* @property y @setter para acceder y modificar el precio (no debe ser menor de 0)

```python
class producto:
    
    impuesto = 0.19  # Atributo de clase
    
    def __init__(self, nombre, precio, codigo):
        self.nombre = nombre
        self._precio = precio
        self.__codigo = codigo
        
    def __eq__(self, other):
        if isinstance(other, producto):
            return self.__codigo == other.__codigo
        return False
    
    @staticmethod
    def aplicar_descuento(precio, porcentaje):
        if 0 <= porcentaje <= 100:
            return precio * (1 - porcentaje / 100)
        else:
            raise ValueError("El porcentaje debe estar entre 0 y 100")
    
    @property
    def precio(self):
        return self._precio
    
    @precio.setter
    def precio(self, modificarPrecio):
        if modificarPrecio < 0:
            raise ValueError("El precio no debe ser menor a 0")
        self._precio = modificarPrecio
        
    def precioConImpuesto(self):
        return self._precio * (1 + producto.impuesto)
    
    def __repr__(self):
        return f"Producto(nombre='{self.nombre}', precio={self._precio}, c√≥digo={self.__codigo})"
    
p1 = producto("arroz", 5000, 100)
p2 = producto("lenteja", 3000, 101)

print(p1) 
print(p2)

# Descuento
print("Precio arroz con 10% de descuento:", producto.aplicar_descuento(p1.precio, 10))

print("Precio con impuesto:", p1.precioConImpuesto())
```

# üßÆ Ejercicio: Clases de Figuras Geom√©tricas
## üéØ Enunciado
### Crea una clase base llamada Figura con las siguientes caracter√≠sticas:

‚úÖ Requisitos:
* Atributos:
* P√∫blico: nombre

* Protegido: _color

* Privado: __id_figura (√∫nico por figura)

### M√©todos:
* M√©todo m√°gico __str__ para mostrar la figura

* M√©todo de clase crear_con_nombre que permite crear figuras por nombre (Figura.crear_con_nombre("Cuadrado"))

* M√©todo est√°tico es_color_valido(color) que retorna True si el color est√° entre ["rojo", "azul", "verde"]

* @property y @setter para acceder y modificar el color (solo si es v√°lido)

### Luego crea una subclase Circulo que herede de Figura:

‚úÖ Requisitos:
### Atributos:
* Protegido: _radio

### M√©todos:
* M√©todo m√°gico __eq__ que compara dos c√≠rculos por radio

* M√©todo area() que devuelva el √°rea del c√≠rculo

* @property y @setter para acceder/modificar el radio (debe ser mayor que 0)

```python
import math
import itertools

class Figura:
    _id_generator = itertools.count(1)  
    
    def __init__(self, nombre, color="rojo"):
        self.nombre = nombre
        self._color = color if self.es_color_valido(color) else "rojo"
        self.__id_figura = next(Figura._id_generator)

    def __str__(self):
        return f"Figura: {self.nombre}, Color: {self._color}, ID: {self.__id_figura}"

    @classmethod
    def crear_con_nombre(cls, nombre):
        return cls(nombre)

    @staticmethod
    def es_color_valido(color):
        return color.lower() in ["rojo", "azul", "verde"]

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, nuevo_color):
        if self.es_color_valido(nuevo_color):
            self._color = nuevo_color
        else:
            print("Color no v√°lido. Usa: rojo, azul o verde.")

class Circulo(Figura):
    def __init__(self, nombre, color="rojo", radio=1):
        super().__init__(nombre, color)
        self._radio = radio if radio > 0 else 1

    def __eq__(self, otro):
        if isinstance(otro, Circulo):
            return self._radio == otro._radio
        return False

    def area(self):
        return math.pi * (self._radio ** 2)

    @property
    def radio(self):
        return self._radio

    @radio.setter
    def radio(self, nuevo_radio):
        if nuevo_radio > 0:
            self._radio = nuevo_radio
        else:
            print("El radio debe ser mayor que 0.")

f1 = Figura.crear_con_nombre("Cuadrado")
print(f1)
f1.color = "azul"
print(f1.color)

c1 = Circulo("C√≠rculo1", "verde", 5)
c2 = Circulo("C√≠rculo2", "rojo", 5)
c3 = Circulo("C√≠rculo3", "azul", 7)

print(c1)
print("√Årea:", c1.area())
print("c1 == c2:", c1 == c2)
print("c1 == c3:", c1 == c3)
```
